# 공격 상황
- Attack: CVE-2016-5195
- 1700249477;1700249479

```python
import matplotlib.pyplot as plt

# Step 1: Read the file
with open('/home/joon/SEMI/Research/Dataset/SocialNetwork-attack-data/scenario1/plain_workload/CVE-2016-5195/collection-1/node1/attack_syscall.txt', 'r') as file:
    lines = file.readlines()

# Step 2: Parse the data and count frequencies
syscall_counts = {}
for line in lines:
    parts = line.strip().split('\t')
    if len(parts) == 2:
        syscall = parts[1]
        if syscall in syscall_counts:
            syscall_counts[syscall] += 1
        else:
            syscall_counts[syscall] = 1

# Step 3: Prepare data for visualization
syscalls = list(syscall_counts.keys())
counts = list(syscall_counts.values())

print(syscalls)
print(counts)
```

```python
# Step 3: Sort the data by frequency and select top 10
sorted_syscalls = sorted(syscall_counts.items(), key=lambda item: item[1], reverse=True)
top_10_syscalls = sorted_syscalls[:10]

# Step 4: Prepare data for visualization
syscalls_10 = [item[0] for item in top_10_syscalls]
counts_10 = [item[1] for item in top_10_syscalls]

# Step 5: Visualize the data
plt.figure(figsize=(10, 5))
plt.bar(syscalls_10, counts_10, color='green')
plt.xlabel('System Call Type')
plt.ylabel('Frequency')
plt.title('Top 10 System Call Types by Frequency')
plt.show()
```
